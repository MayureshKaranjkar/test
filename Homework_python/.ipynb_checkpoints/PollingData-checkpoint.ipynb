````python
#Read .csv
# First we'll import the os module
# This will allow us to create file paths across operating systems
import os

# Module for reading CSV files
import csv
import sys
#import pandas as pd
csvpath = os.path.join('..', 'Resources', 'election_data.csv')

# # Method 1: Plain Reading of CSV files
# with open(csvpath, 'r') as file_handler:
#     lines = file_handler.read()
#     print(lines)
#     print(type(lines))


# Method 2: Improved Reading using CSV module

with open(csvpath, newline='') as csvfile:

    # CSV reader specifies delimiter and variable that holds contents
    csvreader = csv.reader(csvfile, delimiter=',')
    file_to_output = os.path.join("Election_Results_Summary.txt")
    #print(csvreader)

    # Read the header row first (skip this step if there is now header)
    #csv_header = next(csvreader)
    #print(f"CSV Header: {csv_header}")
    #polling_df = pd.read_csv(csvfile)

# Print the first five rows of data to the screen
    #polling_df.head()
    read = csv.reader(csvfile)
    next(read, None)
    
    #Setting original total vote count to 0
    totalvotes = 0
    
    #declaring LIST for getting a list for candidates in this election
    list_candidates_dup=[]
    
    #declaring LIST for getting a list for election results
    #output_file=[]
#output_file.append('Election Results Summary')
#output_file.append('---------------------------------------------')
#print("\n".join((output_file)))  

    #reading candidates and votes in dictionary
    dict = {}
    for row in read:
        totalvotes = totalvotes+1
        list_candidates_dup.append(row[2])
    
    #getting candidate list, if already in list we skip
    #setting new variable list_candidates for each read from dictionary
    list_candidates=set(list_candidates_dup)
    for candidate in list_candidates:
        dict.update({candidate:0})
    
    #counting votes for each candiadte in dictionary
    for dup_candidate in list_candidates_dup:
        if dup_candidate in dict:
            dict[dup_candidate] = dict[dup_candidate]+1
    #print("Total Votes: %s" %totalvotes)
#output_file.append('print("Total Votes: %s" %totalvotes)')    
    ###print("Total Votes: %s" %totalvotes)
    #print(dict)

with open(file_to_output, "w") as txt_file:
    ####print output to file
    election_results = (
        f"-------Election Results-------------\n"
        f"------------------------------------\n"
        f"Total Votes: {totalvotes}\n"
        f"------------------------------------\n"
    )
    print(election_results)
    txt_file.write(election_results)
    
    winner = max(dict,key=dict.get)
    for k,v in dict.items():
        #print(k,round((v*100)/totalvotes,3), v)
        tmp1 = k +": " +str(round((v*100)/totalvotes,3)) +"%    (" + str(v) +")\n"
        print(tmp1)
        txt_file.write(tmp1)
        ###txt_file.write(k,round((v*100)/totalvotes,6), v)
    #print("Winner: %s" %winner)
    tmp2 = "Winner: %s" %winner
    #print(tmp2)
    #txt_file.write(tmp2)
    
    winner_results = (
        f"-----------------------------------\n"
        f" {tmp2}\n"
        f"------------------------------------\n"
    )
    print(winner_results)
    txt_file.write(winner_results)
    #dict_percent.update(dict)
    #print()

#Print output from LIST with appending the values from print stmt
#output_file.append('Election Results Summary')
#output_file.append('---------------------------------------------')
#output_file.append('Total Votes:  ' + str(totalvotes)
#output_file.append('---------------------------------------------')
#output_file.append(print(dict))
#output_file.append('---------------------------------------------')
#output_file.append('Winner: ' + str(winner)
#output_file.append('---------------------------------------------')
#print("\n".join((output_file)))

#Export to .txtfile
#with open("PythonHW2_pyPollingResults.txt", "w") as output:
    #--------------output.write(str(output_file))
    #output.write("\n".join((output_file)))
    
    
  #```text
#  Election Results
 # -------------------------
  #Total Votes: 3521001
  #-------------------------
  #Khan: 63.000% (2218231)
  #Correy: 20.000% (704200)
  #Li: 14.000% (492940)
  #O'Tooley: 3.000% (105630)
  #-------------------------
  #Winner: Khan
  #-------------------------
    
    
````
